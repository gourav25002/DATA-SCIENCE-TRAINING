# **Part 2: The 10 Sales Questions & Answers**

# **Question 1: How many missing (null) values exist in each column of the dataset, and what is the best strategy (dropping vs. imputation) to handle them without losing critical sales data?**

```python
df = pd.read_csv('/content/cleaned_dataset (1).csv')
print("Missing values count:")
print(df.isnull().sum())
```

# **Question 2: Are there case-sensitivity or formatting issues causing duplicate categories in columns like Product_Name, Region, and Payment_Method? How can we programmatically standardize them?**

```python
print("Unique Products:", df['Product_Name'].unique())
print("Unique Regions:", df['Region'].unique())


df['Product_Name'] = df['Product_Name'].str.capitalize()
df['Region'] = df['Region'].str.capitalize()
df['Payment_Method'] = df['Payment_Method'].str.upper()
```

# **Question 3: Are there any rows where the Quantity or Total_Sales columns contain negative numbers or zeros? How do these anomalies affect our overall calculations, and how can we filter them out?**

```python

anomalies = df[df['Quantity'] <= 0]
print(f"Number of rows with invalid quantities: {len(anomalies)}")


df_cleaned = df[df['Quantity'] > 0]
```

# **Question 4: What is the total sales revenue generated by the store, and what is the total volume of items sold across the entire dataset (after filtering out anomalies)?**

```python
total_revenue = df_cleaned['Total_Sales'].sum()
total_items = df_cleaned['Quantity'].sum()

print(f"Total Revenue Generated: ${total_revenue:,.2f}")
print(f"Total Items Sold: {int(total_items)}")
```

# **Question 5: Which product name generates the highest total revenue, and which product is ordered the most frequently in terms of total units sold?**

```python
product_summary = df_cleaned.groupby('Product_Name').agg(
    Total_Units_Sold=('Quantity', 'sum'),
    Total_Revenue=('Total_Sales', 'sum')
).sort_values(by='Total_Revenue', ascending=False)

print(product_summary)
```

# **Question 6: Which geographic region dominates the market in terms of sales revenue, and how does the average transaction value change from region to region?**



```python
region_summary = df_cleaned.groupby('Region')['Total_Sales'].agg(['sum', 'mean', 'count'])
region_summary.columns = ['Total Revenue', 'Average Order Value', 'Order Count']
print(region_summary.sort_values(by='Total Revenue', ascending=False))
```

# **Question 7: Who are the top 5 high-value customers based on their cumulative spending, and how many unique orders did each of them place?**



```python
top_customers = df_cleaned.groupby('Customer_Name')['Total_Sales'].agg(['sum', 'count'])
top_customers.columns = ['Total_Spent', 'Total_Orders']
print(top_customers.sort_values(by='Total_Spent', ascending=False).head(5))
```

# **Question 8: How do sales trend over the recorded months? Which specific month or day of the week experiences peak shopping activity?**

```python

df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'])


df_cleaned['Month'] = df_cleaned['Date'].dt.strftime('%B')

monthly_sales = df_cleaned.groupby('Month')['Total_Sales'].sum().sort_values(ascending=False)
print(monthly_sales)
```

# **Question 9: What is the price of the single most expensive item in the dataset, and what is the price of the cheapest item?**

```python
max_price = df['Unit_Price'].max()
min_price = df['Unit_Price'].min()

print("Most expensive item price:", max_price)
print("Cheapest item price:", min_price)
```
